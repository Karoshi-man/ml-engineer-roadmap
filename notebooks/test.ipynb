```python
import torch
import time
import sys
import os

sys.path.append(os.path.abspath(os.path.join('..')))

from src.device import get_device
```

```python
device = get_device()
```

```python
size = 10000
print(f"Створюємо матриці розміром {size}x{size}...")

start_cpu = time.time()
x_cpu = torch.randn(size, size)
y_cpu = torch.randn(size, size)
z_cpu = torch.mm(x_cpu, y_cpu)
end_cpu = time.time()
print(f"Час на CPU: {end_cpu - start_cpu:.4f} сек")
```

```python
if device.type == 'cuda':
    x_gpu = torch.randn(size, size).to(device)
    y_gpu = torch.randn(size, size).to(device)
    
    _ = torch.mm(x_gpu, y_gpu)
    
    start_gpu = time.time()
    z_gpu = torch.mm(x_gpu, y_gpu)
    torch.cuda.synchronize() 
    end_gpu = time.time()
    
    print(f"Час на GPU: {end_gpu - start_gpu:.4f} сек")
    print(f"Прискорення: {(end_cpu - start_cpu) / (end_gpu - start_gpu):.1f}x")
else:
    print("GPU none.")
```
